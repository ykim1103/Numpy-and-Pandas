#### apply
      - 행이나 열 단위로 더 복잡한 처리를 하고자 할 때 사용
      - 인수로 행 또는 열을 받는 함수를 apply()의 인수로 넣으면 각 열(또는 행)을 반복하여 그 함수에 적용

```python
df=pd.DataFrame({
    'A':[1,3,4,3,4],
    'B':[2,3,1,2,3],
    'C':[1,5,2,4,4,]
})
df

```

```python
def diff(x):
    return x.max()-x.min()

diff(df['A'])  #열이 많아지면 다 이렇게 써주기 힘들기 떄문에 

df.apply(diff)  #이렇게 해준다!
df.apply(diff,axis=1)

df.apply(lambda x:x.max()-x.min())   # 람다 형식으로 쓴 것!
```

```python
print(df['A'].value_counts())
print(df['B'].value_counts())
print(df['C'].value_counts())

#위에 세 개를 apply로 이용해보기
df.apply(pd.value_counts)



#Nan 값을 원하는 값으로 변경: fillna()
df.apply(pd.value_counts).fillna(0).astype(int)
```

#### 실수값을 카테고리값으로 변환
        cut(),qcut()

```python
ages=[0,2,10,21,23,37,31,61,20,41,32,100]

cat=pd.cut(x=ages, bins=[1,15,25,35,60,99],labels=['미성년자','청년','중년','장년','노년'])
print(cat)
print(type(cat))
print(cat.categories)
print(cat.codes)

df=pd.DataFrame(ages,columns=['ages'])
df
df['age_cat']=cat
df
```

```python
# 실습문제

df_score1 = pd.DataFrame({
    "반":["A", "A", "B", "A", "B", "B", "B", "A", "B", "A"],
    "번호" : [1, 2, 1, 3, 2, 3, 4, 4, 5, 5],
    "국어" : [79, 67, 88, 68, 92, 54, 67, 88, 97, 85],
    "영어" : [55, 77, 44, 67, 86, 45, 78, 58, 90, 67],
    "수학" : [57, 45, 76, 68, 89, 67, 99, 78, 89, 90]
})
df_score1


'''1. 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"을 가지는 데이터프레임을 만든다
2. 데이터 프레임에 각 학생의 평균을 나타내는 행을 오른쪽에 추가한다.
3. 행 인덱스로 "번호"를, 1차 열 인덱스로 "국어", "영어", "수학"을, 2차 열 인덱스로 "반"을 가지는 데이터프레임을 만든다.
4. 데이터 프레임에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.
'''
# 1. 1차 행 인덱스로 "반"을 2차 행 인덱스로 "번호"을 가지는 데이터프레임을 만든다

df_score2 = df_score1.set_index(['반', '번호']).sort_index(level = 0, axis = 0)
df_score2


# 2. 데이터 프레임에 각 학생의 평균을 나타내는 행을 오른쪽에 추가한다.
df_score2['평균'] = round(df_score2.mean(axis = 1), 2)
df_score2


# 3. 행 인덱스로 "번호"를, 1차 열 인덱스로 "국어", "영어", "수학"을, 2차 열 인덱스로 "반"을 가지는 데이터프레임을 만든다.

df_score3 = df_score2.unstack('반')
df_score3


# 4. 데이터 프레임에 각 반별 각 과목의 평균을 나타내는 행을 아래에 추가한다.
df_score3.loc['평균', :] = df_score3.mean()
df_score3







#from IPython.display import image
#image('data/img/1.png')
```

#### merge

```python
df1 = pd.DataFrame({
    "고객번호":[1001, 1002, 1003, 1004, 1005, 1006, 1007],
    "이름":["둘리", "도우너", "또치", "길동", "희동", "마이콜", "영희"]
})
print(df1)

df2 = pd.DataFrame({
    "고객번호":[1001, 1001, 1005, 1006, 1008, 1001],
    "금액":[10000, 20000, 15000, 5000, 100000, 30000]
})
print(df2)
```

```python
pd.merge(df1,df2)   # 가장 기본적인 join
```

```python
pd.merge(df1,df2,how='left')
pd.merge(df1,df2,how='right')
pd.merge(df1,df2,how='outer')
```

## 피봇테이블과 그룹분석 : pivot(),groupby(), pivot_table()

##### pivot()
     DataFrame.pivot(self,index=None,columns=None,values=None)

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
df = pd.DataFrame(data)
df
```

```python
# 각 도시에서 연도별 인구수를 알고 싶다. 
df.pivot(index='도시',columns='연도',values='인구')

df.set_index(['도시','연도'])['인구']   #이렇게 하면 시리즈로 나온다.
df.set_index(['도시','연도'])[['인구']] # 인구를 한번 더 묶어줘야 데이터프레임으로 나온다.
df.set_index(['도시','연도'])[['인구']].unstack()
```
