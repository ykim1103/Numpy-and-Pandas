```python
# 각 붓꽃 별로 가장 큰 값과 가장 작은 값의 비율로 구해보자.
def peak_to_peak_ratio(x):
    return x.max()/ x.min()

#############################################

iris.groupby(iris.species).agg(peak_to_peak_ratio)        #1번
print("-------------------------------------------")
iris.groupby(iris.species).apply(peak_to_peak_ratio)      #2번    -> 1번과 같은 결과. 
```

```python
# 각 붓꽃별로 가장 꽃잎길이가 작은 것 3개 데이터를 추출
def peak_to_peak_three(x):
    return x.sort_values(by='petal_length')[:3]       

print(iris.groupby(iris.species).apply(peak_to_peak_three))      

#################################################################

print("-------------------------------------------------------------")
#iris.groupby(iris.species).agg(peak_to_peak_three)     # apply는 되는데 agg로 하면 오류난다. 원인은?
```

```python
def test(x):
    return 0

print(iris.groupby(iris.species).apply(test))
print("--------------------------------------------")
print(iris.groupby(iris.species).agg(test))     #agg는 하나씩 매칭한다?    품종별로 하나씩 다 가져온다?  
```

```python
# 품종별로 합계,평균,표준편차
import numpy as np
iris.groupby('species').agg([np.sum,np.mean,np.std])
#iris.groupby('species').apply([np.sum,np.mean,np.std])    #apply는 오류
```

```python
iris.groupby('species').apply(np.sum)
```

```python
iris.groupby('species').describe().T     # 모든 통계량에 대한 하나하나의 값을 다 처리 . 기초통계를 다 보여준다.!
```

```python
'''
cut: 내가원하는 기준에 맞춰서 자르고
qcut: 일정한게 자른다.
'''

def func_qcut(s):
    return pd.qcut(s,3,labels=['대','중','소'])

###################################################

iris.groupby(iris.species)['petal_length'].transform(func_qcut)

print("-------------------------------------------------------")
iris['petal_length_class']=\
iris.groupby(iris.species)['petal_length'].transform(func_qcut)
iris.head()
```

```python
# 실습예제

df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df



# 도시별로 과일의 가격 평균과 수량 평균을 구해보시오
df.groupby(df.city).mean().T
df.groupby('city').mean().T
# 도시별, 과일별 가격평균과 수량평균을 구해보시오
df.groupby([df.city,df.fruits]).mean().T
df.groupby(['city','fruits']).mean().T
# 위의 문제에서 도시별, 과일별 라벨을 인덱스로 하고싶지 않을 경우
df.groupby([df.city,df.fruits],as_index=False).mean()                

# 도시별로 가격의 평균과 수량의 합계를 동시에 구하고자 했을 때
df.groupby(df.city).agg([np.sum,np.mean]).T                 # 동시에 구할 때는 apply가 아니라 agg를 사용한다. 
df.groupby('city').agg({'price':np.mean, 'quantity':np.sum})
```

#### pivot_table()
        DataFrame.pivot_table(self, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, d                   dropna=True, margins_name='All', observed=False)
        
        - values : 분석할 데이터프레임에서 분석할 열
        - aggfunc : 분석메서드
        - fill_values : NaN 대체값
        - dropna : NaN에 해당하는 값을 버릴지 여부
        - margins : 행과 열 끝에 소계 추가

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.describe()
tips.head()

# tips: 식사 대금 대비 팁의 비율기 어떤 경우에 가장 높아지는지를 찾는 데이터
```

```python
# 식사대금과 팁의 비율을 나타내는  tip_pct를 추가
tips['tip_pct']= tips['tip']/tips['total_bill']*100
tips.head()


# 성별로 인원수 파악
tips.groupby('sex').count()
tips.groupby('sex').size()


# 성별, 흡연 유무별로 인원수 파악
tips.groupby(['sex','smoker']).count()
tips.groupby(['sex','smoker']).size()


# 성별인원수와 성별,흡연유무별로 인원수 파악한 두 정보를 하나의 테이블로 출력
tips.pivot_table(values='tip_pct',index=['sex','smoker'],aggfunc='count',margins=True)
tips.pivot_table('tip_pct','sex','smoker',aggfunc='count',margins=True)

#성별에 따른 팁 비율
tips.pivot_table('tip_pct','sex')
tips.pivot_table(values='tip_pct',index='sex')
tips.groupby('sex')['tip_pct'].mean()       # 시리즈
tips.groupby('sex')[['tip_pct']].mean()     #  데이터프레임


#흡연유무에 따른 팁 비율
tips.groupby('smoker')[['tip_pct']].mean()      # smoker 를 그룹으로 묶은 후 팁비율 평균 보기
tips.pivot_table('tip_pct','sex','smoker')   
tips.pivot_table('tip_pct',['sex','smoker'])   

#팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 분석
tips.pivot_table('tip_pct','day')
tips.pivot_table('tip_pct','time')
tips.pivot_table('tip_pct','size')

tips.pivot_table('tip_pct',['day','time','size'])
tips.groupby(['day','time','size'])[['tip_pct']].mean().dropna()    #dropna() 까지 해줘야 결측치가 사라진다.
```

```python
# 성별,흡연 유무별로 가장 많은 팁과 가장 적은 팁의 차이

def peak_to_peak(x):
    return x.max()-x.min()

tips.groupby(['sex','smoker'])[['tip']].apply(peak_to_peak)

tips.groupby(['sex','smoker']).agg({'tip_pct':'mean','total_bill':peak_to_peak})
```

```python
tips.pivot_table(['tip_pct','size'],['sex','day'],'smoker')   # 성별과 날짜를 기준. 흡연자 유무를 따져서 팁비율과 size출력
```

```python
# 타이타닉
titanic=sns.load_dataset('titanic')
titanic.head()
titanic.count()

#나이를 미성년(1~20),성년(21~60),노년(60~)
#titanic 테이블에 age_class라는 파생변수 추가
bins=[1,20,60,100]
labels=['미성년','성년','노년']
titanic['age_class']= pd.cut(titanic['age'],bins=bins,labels=labels)

#성별,나이대(age_class)별로 객식별 생존여부의 평균값
#titanic.groupby('sex','age_class')[['alive']].mean()

titanic.pivot_table('survived',index=['sex','age_class'],columns='class')
titanic.groupby(['sex','age_class','class'])[['survived']].mean().unstack()
```

```python
data_str=['2020, 1, 1','2020, 1,4','2020, 1, 6','2020, 1, 9']
aa=pd.to_datetime(data_str)
aa


# 날짜를 인덱스로 활용
np.random.seed(0)
s=pd.Series(np.random.randn(4))
s


#날짜 데이터 생성
pd.date_range('2020-1-1','2020-4-30')    #1월1일부터 4월30일까지
pd.date_range('2020-4-1',periods=30)   # 4월1일부터 30일동안의 날짜




"""
freq
    s : 초
    T : 분
    H : 시간
    D : 일
    B : 주말이 아닌 평일
    W : 주(일요일)
    W-MON : 주(월요일)
    M : 각 달의 마지막 날
    MS : 각 달의 첫 날
    BM : 주말이 아닌 평일중에서 각 달의 마지막 날
    BMS : 주말이 아닌 평일 중에서 각 달의 첫날
    WOM-2THU : 각 달의 두번째 목요일
    Q-JAN : 각 분기의 첫 달의 마지막 날
    Q-DEC : 각 분기의 마지막 달의 마지막 날
"""

pd.date_range('2020-1-1','2020-1-30',freq='B')
```

```python
# Shift 연산자
np.random.seed(0)
ts=pd.Series(np.random.randn(4),index=pd.date_range('2020-1-1',periods=4,freq='M'))    #freq='M'  각 달의 마지막 날
ts
print("---------------------------------------------")
print(ts.shift(1))               #shift  의 용도?   데이터나 인덱스를 이동시킬 때 사용. 
print(ts.shift(-1))      
print(ts.shift(1,freq='M'))
print(ts.shift(1,freq='w'))      
```

```python
# resampling : 시간간격을 다시 재조정
'''
up-sampling : 시간 구간이 작아지면서 데이터 양은 증가
down-sampling : 구간이 커지면서 데이터 양의 감소
'''


#down-sampling 
ts=pd.Series(np.random.randn(100),index=pd.date_range('2020-1-1',periods=100))
ts.head(10)

print("______________________________")
ts.resample("w").mean()
ts.resample('M').first()


#up-sampling
'''
ffill() : forward filling : 앞에서 나온 데이터를 그대로 사용
bfill() : backwardfilling : 뒤에서 나올 데이터를 앞에서 미리 사용

'''
ts.resample('30s').ffill().head(20)
ts.resample('30s').bfill().head(20)

```
